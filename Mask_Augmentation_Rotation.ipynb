```python
from PIL import Image
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
input_folder = '/content/drive/MyDrive/Skripsi/Data/Mask/Mask'
output_folder = '/content/drive/MyDrive/Skripsi/Data/Mask/Mask_Rot180'

os.makedirs(output_folder, exist_ok=True)
```

```python
def rotate_180(image):
    return image.rotate(180, expand=True)

displayed = False

for filename in os.listdir(input_folder):
    if filename.lower().endswith('.jpg'):
        basename = os.path.splitext(filename)[0]

        img = Image.open(os.path.join(input_folder, filename))
        img_rot180 = rotate_180(img)

        # Menyimpan hasil rotasi 180
        img_rot180.save(os.path.join(output_folder, f"{basename}_rot180.jpg"))

        # Menampilkan contoh hasil augmentasi
        if not displayed:
            plt.figure(figsize=(10, 4))
            plt.subplot(1, 2, 1)
            plt.imshow(img, cmap='gray')
            plt.title("Original")
            plt.axis("off")

            plt.subplot(1, 2, 2)
            plt.imshow(img_rot180, cmap='gray')
            plt.title("Vertical Flip")
            plt.axis("off")

            plt.suptitle(f"Augmentasi untuk {filename}")
            plt.tight_layout()
            plt.show()

            displayed = True
```
