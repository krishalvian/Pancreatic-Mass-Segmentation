```python
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
input_folder = '/content/drive/MyDrive/Skripsi/Data/Mask/Mask'
output_folder = '/content/drive/MyDrive/Skripsi/Data/Mask/Mask_Hflip'

os.makedirs(output_folder, exist_ok=True)
```

```python
def horizontal_flip(image):
    return image.transpose(Image.FLIP_LEFT_RIGHT)

# Flag untuk menampilkan hanya satu contoh
displayed = False

for filename in os.listdir(input_folder):
    if filename.lower().endswith('.jpg'):
        basename = os.path.splitext(filename)[0]

        img = Image.open(os.path.join(input_folder, filename))

        img_hflip = horizontal_flip(img)

        # Save hasil augmentasi
        img_hflip.save(os.path.join(output_folder, f"{basename}_hflip.jpg"))

        # Tampilkan hanya satu contoh
        if not displayed:
            plt.figure(figsize=(10, 4))
            plt.subplot(1, 2, 1)
            plt.imshow(img, cmap='gray')  # cmap='gray' jika ini citra mask biner
            plt.title("Original")
            plt.axis("off")

            plt.subplot(1, 2, 2)
            plt.imshow(img_hflip, cmap='gray')
            plt.title("Horizontal Flip")
            plt.axis("off")

            plt.suptitle(f"Augmentasi untuk {filename}")
            plt.tight_layout()
            plt.show()

            displayed = True
```
