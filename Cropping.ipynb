Mengimpor modul PIL (Python Imaging Library) untuk bekerja dengan gambar

```python
from PIL import Image
import matplotlib.pyplot as plt
```

Menentukan gambar yang ingin dilakukan *cropping*

```python
image_path = "/content/drive/MyDrive/Dataset/Raw/Data/train/Raw/0840756_Rep#2#6.jpg"
image = Image.open(image_path)
```

```python
# Define crop area (left, top, right, bottom)
left = 170   # Adjust these values as needed
top = 80
right = 550
bottom = 410

# Crop the image
cropped_image = image.crop((left, top, right, bottom))

# Save the cropped image
cropped_image_path = "/content/drive/MyDrive/Dataset/Raw/Data/train/Raw/#01cropped.jpg"
cropped_image.save(cropped_image_path)
```

```python
# Show the cropped image
plt.imshow(cropped_image)
plt.axis("off")
plt.show()
```

Gambar diatas merupakan hasil dari proses *cropping*
