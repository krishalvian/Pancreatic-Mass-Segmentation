```python
import os

def get_filenames(folder):
    # Ambil nama file tanpa ekstensi
    return {os.path.splitext(f)[0] for f in os.listdir(folder) if os.path.isfile(os.path.join(folder, f))}

# Ganti path folder sesuai kebutuhanmu
folder1 = "/content/drive/MyDrive/Final/Dataset/Raw"  # Misal: folder images
folder2 = "/content/drive/MyDrive/Final/Dataset/Mask"  # Misal: folder masks

# Ambil nama file (tanpa ekstensi) dari kedua folder
files1 = get_filenames(folder1)
files2 = get_filenames(folder2)

# Cek kesamaan
only_in_1 = files1 - files2
only_in_2 = files2 - files1
common = files1 & files2

# Tampilkan hasil
print(f"Total file di Folder 1: {len(files1)}")
print(f"Total file di Folder 2: {len(files2)}")
print(f"Jumlah file yang match (nama sama): {len(common)}")

if only_in_1:
    print(f"\nFile yang hanya ada di Folder 1 ({len(only_in_1)}):")
    for f in sorted(only_in_1): print(f"  {f}")
if only_in_2:
    print(f"\nFile yang hanya ada di Folder 2 ({len(only_in_2)}):")
    for f in sorted(only_in_2): print(f"  {f}")

if not only_in_1 and not only_in_2:
    print("\n✅ Semua file match berdasarkan nama.")
else:
    print("\n❌ Ada file yang tidak punya pasangan.")
```
