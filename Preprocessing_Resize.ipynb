```python
import os
import cv2
import numpy as np
```

```python
folder_path = '/content/drive/MyDrive/Final'
```

```python
heights = []
widths = []

for filename in os.listdir(folder_path):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        img_path = os.path.join(folder_path, filename)
        img = cv2.imread(img_path)
        if img is not None:
            h, w = img.shape[:2]
            heights.append(h)
            widths.append(w)
            print(f"{filename}: {img.shape}")
        else:
            print(f"{filename}: Gagal dibaca")

if heights and widths:
    avg_height = np.mean(heights)
    avg_width = np.mean(widths)
    print(f"\nRata-rata shape: ({avg_height:.2f}, {avg_width:.2f})")
else:
    print("Tidak ada gambar yang berhasil dibaca.")
```

```python
# Path input dan output folder
input_folder = '/content/drive/MyDrive/Final'
output_folder = '/content/drive/MyDrive/Final/Resize'  # Output di folder terpisah

os.makedirs(output_folder, exist_ok=True)

# Target size (width, height) untuk melakukan resize
target_size = (1520, 1320)

# Looping untuk seluruh image
for filename in os.listdir(input_folder):
    if filename.lower().endswith('.jpg'):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        img = cv2.imread(input_path)

        if img is not None:
            resized = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)
            cv2.imwrite(output_path, resized)
            print(f"{filename}: {resized.shape}")
        else:
            print(f"Gagal dibaca: {filename}")
```
