```python
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
input_folder = '/content/drive/MyDrive/Skripsi/Data/Raw/Raw'
output_folder = '/content/drive/MyDrive/Skripsi/Data/Raw/Raw_Rot180'

os.makedirs(output_folder, exist_ok=True)
```

```python
def rotate_180(image):
    return image.rotate(180, expand=True)

displayed = False  # hanya tampilkan satu kali

for filename in os.listdir(input_folder):
    if filename.lower().endswith('.jpg'):
        basename = os.path.splitext(filename)[0]

        img = Image.open(os.path.join(input_folder, filename))
        img_rot180 = rotate_180(img)

        # Simpan hasil rotasi 180
        img_rot180.save(os.path.join(output_folder, f"{basename}_rot180.jpg"))

        # Tampilkan hanya sekali
        if not displayed:
            plt.figure(figsize=(10, 4))
            plt.subplot(1, 2, 1)
            plt.imshow(img)
            plt.title("Original")
            plt.axis("off")

            plt.subplot(1, 2, 2)
            plt.imshow(img_rot180)
            plt.title("Rotate 180Â°")
            plt.axis("off")

            plt.suptitle(f"Augmentasi untuk {filename}")
            plt.tight_layout()
            plt.show()

            displayed = True  # tandai sudah ditampilkan
```
