```python
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
input_folder = '/content/drive/MyDrive/Skripsi/Data/Mask/Mask'
output_folder = '/content/drive/MyDrive/Skripsi/Data/Mask/Mask_Vflip'

os.makedirs(output_folder, exist_ok=True)
```

```python
def vertical_flip(image):
    return image.transpose(Image.FLIP_TOP_BOTTOM)

displayed = False

for filename in os.listdir(input_folder):
    if filename.lower().endswith('.jpg'):
        basename = os.path.splitext(filename)[0]

        img = Image.open(os.path.join(input_folder, filename))

        img_vflip = vertical_flip(img)

        # Menyimpan hasil augmentasi
        img_vflip.save(os.path.join(output_folder, f"{basename}_vflip.jpg"))

        # Menampilkan contoh hasil augmentasi
        if not displayed:
            plt.figure(figsize=(10, 4))
            plt.subplot(1, 2, 1)
            plt.imshow(img, cmap='gray')
            plt.title("Original")
            plt.axis("off")

            plt.subplot(1, 2, 2)
            plt.imshow(img_vflip, cmap='gray')
            plt.title("Vertical Flip")
            plt.axis("off")

            plt.suptitle(f"Augmentasi untuk {filename}")
            plt.tight_layout()
            plt.show()

            displayed = True
```
